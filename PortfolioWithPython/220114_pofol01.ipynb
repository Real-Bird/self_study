# 단리이자 계산


*   원금 = 1,000만 원
*   이율 = (연간) 5%
*   기간 = 1년
*   1년 후 인출액 = 원금 * (1 + 이율 * 기간)





```python
# 원금
a = 1000
# 기간
n = 1
# 연이율
r = 0.05
# 단리계산 식
s_simple = a * (1 + r * n)

print(s_simple)
```

```python
import scipy as sp

s_simple = sp.fv(r, n, 0, a)
print(s_simple)
```

# 복리계산


*   1년 후 인출액 = 원금 * (1 + (연이율 / 12개월))**12개월



```python
# 원금
a = 1000
# 기간(개월)
n = 12
# 연이율
r = 0.05
# 복리계산 식
s_compound = a * (1 + (r / n)) ** n

print(s_compound)
```

# NPV
- 순현재가치(Net Present Value)
- NPV = SUM(현금흐름/(1+할인율)**기간)

```python
# 현금흐름
cashflows = [12000, 15000, 18000, 21000, 26000]
# 햇수
i = 0
# 이자율
r = 0.015

# 최초 투자금액이며 현금 유출이므로 (-)로 표시
npv = -70_000

for c in cashflows:
  i += 1
  npv += c / (1 + r) ** i

print(npv)
```

```python
npv = sp.npv(r, cashflows)
print(npv)
```

# IRR
- 내부수익률(Internal Rate of Return)

```python
cashflows = [-70000, 12000, 15000, 18000, 21000, 26000]

irr = numpy_financial.irr(cashflows)
npv = numpy_financial.npv(irr, cashflows)
print('IRR {0:.1%} makes NPV {1:.2f}'.format(irr, npv))
```

# 산술평균, 기하평균

```python
# 기간별 수익률
returns = [0.1, 0.06, 0.05]
# 합계
sumOfReturns = 0.0
# 산술평균, 기하평균
arimean = 0.0
geomean = 1.0

# 기간별 수익률 데이터 개수
n = len(returns)

# 수익률 합계
for r in returns:
  sumOfReturns += r

# 산술평균 출력
arimean = sumOfReturns / n
print('AriMean is {:.2%}'.format(arimean))

# 기하평균 출력
for r in returns:
  geomean += (1 + r)
# 기간 수익률로 변환
geomean **= (1 / n) - 1
print('GeoMean is {:.2%}'.format(geomean))
```
