```python
%%timeit

c = np.zeros((size, size))

for i in range(size):
    if i % 100 == 0:
        print("i = ", i)
    for j in range(size):
        for k in range(size):
            c[i, j] += a[i, k] * b[k, j]
        pass
    if i == size - 1:
        break
```
